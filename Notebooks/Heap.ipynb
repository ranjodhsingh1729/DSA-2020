```python
from queue import Queue
from graphviz import Digraph

"""
Heap Abstract Data Type,
Implementation in Python.
"""

class Heap:
    def __init__(self):
        self.items = []
    
    
    @staticmethod
    def left(index):
        return 2*index+1
    
    
    @staticmethod
    def right(index):
        return 2*index+2
    
    
    def is_leaf(self, index):
        length = len(self)
        
        if index > length//2:
            return True
        
        else:
            return False
    
    
    def size(self):
        return len(self.items)
    
    
    def visualize(self, index=0):
        graph = Digraph()
        size = self.size()
        remaining = Queue(0)
        remaining.put(index)
        
        while not remaining.empty():
            current = remaining.get()
            L = self.left(current)
            R = self.right(current)
            
            if L < size:
                remaining.put(L)
                graph.edge(str(self.items[current]), str(self.items[L]))
            if R < size:
                remaining.put(R)
                graph.edge(str(self.items[current]), str(self.items[R]))
        
        return graph
    
    
    def heapify(self, index, size):
        largest = index # Initialising Largest to Root
        L = self.left(index)  # Index of Left Node
        R = self.right(index) # Index of Right Node
        
        if L < size and self.items[L] > self.items[largest]:
            print()
            largest = L
        
        if R < size and self.items[R] > self.items[largest]:
            largest = R
        
        if largest != index:
            self.items[index], self.items[largest] = self.items[largest], self.items[index]
            self.heapify(largest, size)
    
    
    def build_heap(self):
        size = self.size()
        
        for i in reversed(range(size//2)):
            self.heapify(i, size)
    
    def sort_heap(self):
        size = self.size()
        
        self.build_heap()
        
        for i in reversed(range(size)):
            self.items[i], self.items[0] = self.items[0], self.items[i]
            self.heapify(0, i)
```

```python
heap = Heap()
heap.items = [100, 5, 3, 2, 8, 15, 6, 102]
```
